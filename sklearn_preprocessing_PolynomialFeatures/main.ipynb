```python
import numpy as np; np.random.seed(42)

m = 100
x = 10 * np.random.uniform(size=(m, 1)) - 3
y = 0.7 * x**2 + x + 1 + np.random.normal(size=(m, 1))

print(x.shape, y.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x, y, "b.")
plt.xlabel("$x$")
plt.ylabel("$y$")

plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

a = PolynomialFeatures(degree=2, include_bias=False)
x_ = a.fit_transform(x)

print(x.shape, x_.shape)
print(x[0], x_[0], x[0]**2)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_, y)
y_pred = lin_reg.predict(x_)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x, y, "b.")
plt.xlabel("$x$")
plt.ylabel("$y$")

print(x.shape, y_pred.shape)
idx = np.argsort(x.reshape((-1,)))
plt.plot(x.reshape((-1,))[idx], y_pred.reshape((-1,))[idx])

plt.show()
```
